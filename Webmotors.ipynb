** Criação das Tabelas e Colunas por Variaveis - Dim_Anuncio  **

```python

Dim_Anuncio = pd.DataFrame({
    
    "cod_anuncio" : dados["cod_anuncio"], 
    "prioridade" : dados["prioridade"], 
    "cep_2dig" : dados["cep_2dig"], 
    "uf_cidade" : dados["uf_cidade"]
})

Dim_Anuncio
```

```python
## Mapeando as letras com acentos 
gregas_acentuadas = {'Ã£' : 'ã',
                     'Ã©' : 'e', 
                     'Ã³' : 'o', 
                     'Ã§' : 'ç',
                     'Ãº' : 'u', 
                     'Ã'  : 'i', 
                     'iª' : 'ê', 
                     'i¢' : 'â', 
                     'i¡' : 'á'} 

## Corrigindo as letras gregas 
for grega, acentuada in gregas_acentuadas.items():
    
    Dim_Anuncio['uf_cidade'] = Dim_Anuncio['uf_cidade'].str.replace(grega, acentuada)
```

```python
Dim_Anuncio['uf_cidade'].sample(20)
```

```python
Dim_Anuncio
```

```python
Dim_Anuncio[['Estado', 'Cidade']] = Dim_Anuncio['uf_cidade'].str.split('_', expand=True)
```

```python
Dim_Anuncio = Dim_Anuncio.drop(columns=['uf_cidade'])
```

```python
## Verificação de dados nulos
Dim_Anuncio.isna().sum()
```

```python
## Verificação de dados duplicados
Dim_Anuncio.duplicated().sum()
```

```python
Dim_Anuncio.to_csv("anuncio.csv")
```
